```python
import pandas as pd

data_gpu = {
    'Model': ['RTX 3060', 'RX 6600', 'RTX 5060 Ti', 'Arc A750'],
    'Brand': ['NVIDIA', 'AMD', 'NVIDIA', 'Intel'],
    'Price_IDR': [4500000, 3800000, 7500000, 3500000],
    'VRAM_GB': [12, 8, 16, 8]
}
```

```python
df = pd.DataFrame(data_gpu)
```

```python
print("--- Tabel GPU Lengkap ---")
print(df)
```

```python
vram_besar = df[df['VRAM_GB'] > 8]
print(vram_besar)
```

```python
rata_harga = df.groupby('Brand')['Price_IDR'].mean()
print(rata_harga)
```

```python
ringkasan = df.groupby('Brand').agg({
  'Price_IDR': ['min','max','mean'],
  'Model': 'count'
})
print(ringkasan)
```

```python
df_urut = df.sort_values(by='Price_IDR')
print(df_urut)
```

```python
df_urut = df.sort_values(by='Price_IDR', ascending=False)
print(df_urut)
```
