```python
import pandas as pd
```

```python
data_transport = {
    'Moda': ['MRT', 'MRT', 'TransJakarta', 'TransJakarta', 'LRT', 'LRT'],
    'Rute': ['Lebak Bulus-HI', 'HI-Lebak Bulus', 'Blok M-Kota', 'Ragunan-Dukuh Atas', 'Pegangsaan-Velodrome', 'Velodrome-Pegangsaan'],
    'Penumpang_Harian': [85000, 82000, 120000, 95000, 5000, 4800],
    'Tarif_Max': [14000, 14000, 3500, 3500, 5000, 5000]
}

df_jkt = pd.DataFrame(data_transport)
```

```python
# --- TANTANGAN ---
# 1. Coba hitung RATA-RATA Penumpang_Harian per 'Moda'
# 2. Pakai .agg() buat cari Tarif_Max paling murah ('min') dan paling mahal ('max') per 'Moda'

print("--- Data Transportasi Jakarta ---")
print(df_jkt)
```

```python
rata_penumpang = df_jkt.groupby('Moda')['Penumpang_Harian'].mean()
print(rata_penumpang)
```

```python
tarif_murah_mahal = df_jkt.groupby('Moda').agg({
  'Tarif_Max': ['min','max']
})
print(tarif_murah_mahal)
```
