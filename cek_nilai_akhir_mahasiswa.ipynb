```python
data_mahasiswa = [
    {"nama": "Andi", "tugas": 80, "uts": 75, "uas": 85},
    {"nama": "Budi", "tugas": 60, "uts": 50, "uas": 40},
    {"nama": "Cici", "tugas": 90, "uts": 90, "uas": 95},
    {"nama": "Dedi", "tugas": 50, "uts": 40, "uas": 30}
    #Tambah terus kalau butuh banyak
]
```

```python
def hitung_nilai_akhir(tugas,uts,uas):
  final_tugas = tugas * 0.3
  final_uts = uts * 0.3
  final_uas = uas * 0.4
  nilai_akhir = final_tugas + final_uas + final_uts
  return nilai_akhir
  
```

```python
def tentukan_grade(nilai):
  if nilai >= 85:
    grade = "A"
  elif nilai >= 70:
    grade = "B"
  elif nilai >= 55:
    grade = "C"
  elif nilai >= 40:
    grade = "D"
  else:
    grade = "E"
  return grade
```

```python
def proses_mahasiswa(data_mahasiswa):
  for mahasiswa in data_mahasiswa:
    nama_mahasiswa = mahasiswa["nama"]
    tugas_mahasiswa = mahasiswa["tugas"]
    uts_mahasiswa = mahasiswa["uts"]
    uas_mahasiswa = mahasiswa["uas"]
    nilai_akhir = hitung_nilai_akhir(tugas_mahasiswa,uts_mahasiswa,uas_mahasiswa)
    grade = tentukan_grade(nilai_akhir)
    print("Nama: " + nama_mahasiswa + " , Nilai Akhir: " + str(nilai_akhir) + " , Grade: " + grade)
```

```python
proses_mahasiswa(data_mahasiswa)
```
