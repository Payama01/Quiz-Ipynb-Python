```python
import numpy as np
```

```python
# Soal 1: Operasi Matematika Element-wise (Diskon GPU)
# Kamu sedang memantau harga beberapa kartu grafis (GPU) rakitan PC dalam satuan juta Rupiah.
harga_gpu = [5.5, 7.2, 8.0, 12.5, 4.8]
#Tugasmu:
# Ubah list harga_gpu menjadi NumPy array.
# Toko sedang mengadakan Flash Sale diskon 15% untuk semua GPU. Artinya, kamu bisa mendapatkan harga baru dengan mengalikan harga lama dengan 0.85. Buat variabel harga_diskon menggunakan operasi matematika langsung ke array tersebut.
# Cetak harga GPU yang paling murah setelah diskon menggunakan fungsi statistik NumPy.
```

```python
harga_gpu = np.array(harga_gpu)
harga_diskon = harga_gpu * 0.85
print("Harga GPU setelah diskon:", harga_diskon)
print("Harga GPU termurah: {}".format(harga_diskon.min()))
```

```python
# Soal 2: Boolean Array & Perhitungan (Monitoring Server)
# Kamu sedang mengawasi respon waktu (ping) dari sebuah server Linux dalam satuan milidetik (ms).
ping_server = [45, 120, 30, 250, 55, 80, 105, 15]
# Tugasmu:
# Ubah list ping_server menjadi NumPy array.
# Server dianggap "Ngelag" jika ping berada di atas 100 ms. Buat array Boolean untuk mengecek kondisi ini.
# Gunakan trik .sum() untuk menghitung secara total ada berapa kali server tersebut mengalami lag. Cetak angkanya!
```

```python
ping_server = np.array(ping_server)
boolean_ping = ping_server > 100
print("Total Server Mengalami Lag: {}".format(boolean_ping.sum())) 
```

```python
# Soal 3: Menggabungkan Statistik & Persentase (Suhu CPU)
# Kamu merekam suhu prosesor saat sedang melakukan stress test (dalam derajat Celcius).
suhu_cpu = [45, 50, 55, 85, 90, 48, 52, 88]
# Tugasmu:
# Ubah list menjadi NumPy array.
# Cari nilai rata-rata (mean) dari seluruh suhu tersebut dan simpan di variabel rata_rata_suhu.
# Kamu ingin tahu seberapa sering CPU ini kepanasan di atas normal. Buat filter Boolean untuk mencari suhu yang lebih tinggi dari rata_rata_suhu.
# Hitung persentase berapa kali CPU tersebut melewati batas rata-ratanya (jumlah yang kepanasan dibagi total data yang direkam). Cetak hasil persentasenya!
```

```python
suhu_cpu = np.array(suhu_cpu)
mean_suhu = suhu_cpu.mean()
boolean_suhu = suhu_cpu > mean_suhu
persen_suhu = boolean_suhu.sum() / len(suhu_cpu) *100
print("Suhu Rata-rata: {}".format(mean_suhu))
print("berapa kali suhu melewati batas rata-rata: {}".format(boolean_suhu.sum()))
print("Persentase Suhu yang lewat batas rata-rata: {}%".format(persen_suhu))
```
