```python
#Soal 1
#Membuat Dictionary dari Dua List (The Zip)
#Kamu punya dua list data penjualan motor bulan ini.
merk = ["Honda", "Yamaha", "Suzuki", "Kawasaki"]
terjual = [150, 120, 90, 60]
```

```python
data_penjualan = dict(zip(merk,terjual))
print(data_penjualan)
```

```python
#Soal 2
#Manipulasi Dictionary (Add & Update)
#Lanjutkan dengan dictionary data_penjualan dari Soal 1.
# Ternyata ada revisi data:
# Tugasmu:
# Penjualan "Honda" ternyata salah catat, ubah nilainya menjadi 180.
# Ada merk baru masuk, tambahkan "Vespa" dengan penjualan 45 ke dalam dictionary.
# Cetak dictionary yang sudah diperbarui.
data_penjualan["Honda"] = 180
data_penjualan["Vespa"] = 45
print(data_penjualan)
```

```python
# Soal 3: Filtering Dictionary (Looping & If)
# Manajer ingin tahu merk apa saja yang penjualannya di atas 100 unit.
# Tugasmu:
# Buat loop for untuk mengakses key (merk) dan value (jumlah) dari data_penjualan (Gunakan .items()).
# Di dalam loop, gunakan if untuk mengecek apakah jumlah terjual lebih besar dari 100.
# Jika ya, cetak nama merknya.
print("Jumlah Unit terjual di atas 100:")
for merk, jumlah in data_penjualan.items():
    if jumlah > 100:
        print("Merek: {} memiliki penjualan sebesar {} unit".format(merk,jumlah))
```

```python
# Soal 4: Membersihkan Duplikat dengan Set
# Kamu melakukan survei motor apa yang lewat di jalan, dan hasilnya ada banyak nama yang sama (duplikat).
survey_jalan = ["Beat", "Vario", "Beat", "Mio", "Vario", "Nmax", "Beat"]
```

```python
motor_unik = set(survey_jalan)
print("Motor yang unik adalah: {}".format(motor_unik))
```

```python
# Soal 5: Operasi Matematika Set (Intersection)
# Kamu punya data stok dari dua gudang berbeda.
gudang_A = {"Oli", "Ban", "Kampas Rem", "Busi"}
gudang_B = {"Ban", "Spion", "Busi", "Lampu"}
# Cari barang apa saja yang ADA DI KEDUA gudang tersebut (barang yang sama/irisan).
# Gunakan metode .intersection() atau operator &.
# Cetak hasilnya.
```
