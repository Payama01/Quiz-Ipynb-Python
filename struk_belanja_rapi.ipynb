```python
barang = "kopi"
harga = 12500.75
jumlah = 3
```

```python
def total_harga(barang,harga,jumlah):
  harga_total = jumlah * harga
  print("Anda membeli {} {}." .format(jumlah,barang))
  print("Harga satuan: Rp {:.2f}" .format(harga))
  print("Total bayar: Rp {:.2f}" .format(harga_total))
  return
```

```python
total_harga(barang,harga,jumlah)
```
